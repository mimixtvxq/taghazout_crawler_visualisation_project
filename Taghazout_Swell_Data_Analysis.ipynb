```python
# Read CSV File 
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

filename = 'taghazout_surf_data.csv'

df = pd.read_csv(filename, parse_dates = True)

print df.columns
```

```python
#convert date into datetime format
from datetime import datetime
import matplotlib.dates as mdates
dates = df["date"]
#type(dates)
#print dates
df["times"] = [datetime.strptime(d, '%Y/%d/%m') for d in dates]
```

```python
#extract years
df["year"] = pd.DatetimeIndex(df["times"]).year
df["month"] = pd.DatetimeIndex(df["times"]).month

print df.head()
```

```python
#convert feet to metres and attach it to df
swell_height = []
for i in range(0,len(df["date"])):
    swell_height.append(df["swell_height_ft"][i]*0.3048)
#print swell_height

#attach conversion to df
df["swell_height"] = swell_height

print df.head()
```

```python
import os
import imageio

png_dir = 'Images_World'
images = []
for file_name in os.listdir(png_dir):
    if file_name.endswith('.png'):
        file_path = os.path.join(png_dir, file_name)
        images.append(imageio.imread(file_path))
imageio.mimsave('gif/global_swell_height.gif', images, duration = 0.5)
```
